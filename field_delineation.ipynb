```python
import cv2
import numpy as np

img = cv2.imread("test.png")
# create engine
engine = cv2.hfs.HfsSegment_create(img.shape[0], img.shape[1])
engine.setSlicSpixelSize(8)
# perform segmentation
# now "res" is a matrix of indices
# change the second parameter to "True" to get a rgb image for "res"
res = engine.performSegmentCpu(img, True)
```

```python
from matplotlib.pyplot import imshow
imshow(res)
```

```python
mask = cv2.compare(res, 17, cv2.CMP_EQ)  # 0/1 mask for blob 17, same size as input image
```

```python
cnt = cv2.findContours(mask,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE, offset=(2,2))
```

```python
contours, hierarchy = cv2.findContours(image=res, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)

# draw contours on the original image
image_copy = img.copy()
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
   
```
